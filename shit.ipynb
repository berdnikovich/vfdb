```python
def id_from_head(h):
    return h.split(" ")[0][1:]
```

```python
with open('table_free_info_ecolivf_vfdb', 'w') as table:
    with open(q, "r") as qu, open(db, "r") as bd:
        parserq = fastaparser.Reader(qu, parse_method='quick')
        parserdb = fastaparser.Reader(bd, parse_method='quick')
        for index, row in blastn_file_ecolivsvfbd.iterrows():
            i_eco, i_vf = row["query id"], row["target id"] 
            for seqq in parserq:
            # seq is a namedtuple('Fasta', ['header', 'sequence'])
                qhead = seqq.header
                qid = id_from_head(qhead)
                if qid == i_eco:
                    table.write(qhead +"\t")#+dbhead+"\n")
                    break                    
            for seqdb in parserdb:
            # seq is a namedtuple('Fasta', ['header', 'sequence'])
                dbhead = seqdb.header
                dbid = id_from_head(dbhead)
                if dbid == i_vf:
                    table.write(dbhead +"\n")#+dbhead+"\n")
                    break                    
            
```
